## Waffle Chart

```python
# Create new DataFrame for countries
df_dan=df_can.loc[['Denmark','Norway','Sweden'],:]

df_dan
```

```python
# cpmpute the proportion of each category with respect to total
total_values=sum(df_dan['Total'])
category_proportions=[(float(value)/total_values)for value in df_dan['Total']]

#Print proportions
for i,proportion in enumerate(category_proportions):
    print(df_dan.index.values[i]+':'+ str(proportion))
```

```python
# Overall size of waffle chart

width=40
height=10

#Total tiles
total_num_tiles=width * height 

print('Total number of tiles is ',total_num_tiles)
```

```python
# Proprtion of each category

tiles_per_category=[round(proportion * total_num_tiles) for proportion in category_proportions]

#Print Number of tiles
for i,tiles in enumerate(tiles_per_category):
    print(df_dan.index.values[i] + ':' + str(tiles))
```

```python
# Creating  matrix

waffle_chart=np.zeros((height,width))

#define indices to loop through waffle chart
category_index=0
tile_index=0

#Populate waffle chart
for col in range(width):
    for row in range(height):
        tile_index +=1
        
        # if no. of tiles populated for current category
        if tile_index > sum(tiles_per_category[0:category_index]):
            category_index +=1
            
        #set class value to an integer which increase class
        waffle_chart[row,col]=category_index
        
print('Waffle chart Populated!')
```

```python
waffle_chart
```

```python
## Map the waffle chart in matrix visual

# initiate a new fig object
fig=plt.figure()

# Use matshow to display waffle chart

colormap=plt.cm.coolwarm
plt.matshow(waffle_chart,cmap=colormap)
plt.colorbar()
```

## Prettify Chart

```python
# initiate a new fig object
fig=plt.figure()

# Use matshow to display waffle chart

colormap=plt.cm.coolwarm
plt.matshow(waffle_chart,cmap=colormap)
plt.colorbar()

#get axis
ax=plt.gca()

#set Minor ticks
ax.set_xticks(np.arange(-.5,(width),1),minor=True)
ax.set_yticks(np.arange(-.5,(height),1),minor=True)

#Add gridlines
ax.grid(which='minor',color='w',linestyle='-',linewidth=2)

plt.xticks([])
plt.yticks([])
```

# Word Clouds

```python
# import package and its set of stopwords
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS

print('Wordcloud is installed and imported!')
```

```python
# open file and read alice_novel

alice=open('alice.txt','r').read()

print('File downloaded and save')
```

```python
stopwords=set(STOPWORDS)
```

```python
# instantiate a word cloud object
alice_wc=WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# Generate WC
alice_wc.generate(alice)
```

```python
#Display WC
plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig=plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')

alice_wc.generate(alice)

fig=plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
